<b>Project Data Mahasiswa</b><br>
Sampel 1000 data mahasiswa, beragam memiliki 10 kolom<br>
<ol>
    <li>Usia         : 18 - 22</li>
    <li>Gender       : L/P</li>
    <li>Jurusan      : TI, TS, TE, TM</li>
    <li>Asal Sekolah : SMA, SMK, MA</li>
    <li>Semester     : 1, 2, 3, 4, 5, 6, 7, 8</li>
    <li>Kelas        : Reguler, Karyawan, Blended</li>
    <li>Mata Kuliah  : Matematika, Fisika, kimia</li>
    <li>Absen        : Hadir, Izin, Sakit, Alfa</li>
    <li>IPK          : 2, 3, 4</li>
    <li>Kategori     : Beasiswa, NonBeasiswa</li>
</ol>

```python
import numpy as np
import pandas as pd

n_data = 1000

usia = np.random.randint(18,23,size=n_data)
gender = np.random.choice(['L','P'],size=n_data)
jurusan = np.random.choice(['TI','TS','TE','TM'],size=n_data)
asalSekolah = np.random.choice(['SMA','SMK','MA'],size=n_data)
semester = np.random.randint(1,9,size=n_data)
kelas = np.random.choice(['Reguler','Karyawan','Blended'],size=n_data)
matkul = np.random.choice(['MTK','Fisika','Kimia'],size=n_data)
absen = np.random.choice(['Hadir','Izin','Sakit','Alfa'],size=n_data)
ipk = np.random.randint(2,5,size=n_data)
kategori = np.random.choice(['Beasiswa','Non Beasiswa'],size=n_data)

df = pd.DataFrame({
    'Usia' : usia,
    'Gender' : gender,
    'Jurusan' : jurusan,
    'Asal Sekolah' : asalSekolah,
    'Semester' : semester,
    'Kelas' : kelas,
    'Mata Kuliah' : matkul,
    'Absen' : absen,
    'IPK' : ipk,
    'Kategori' : kategori
})

print(df.info)
print(f"Usia mahasiswa termuda: {df.Usia.min()}")
print(f"Usia mahasiswa tertua: {df.Usia.max()}")
jmlL = df[df['Gender']=='L']
jmlP = df[df['Gender']=='P']
print(f"Jumlah mahasiswa laki-laki: {jmlL.size/10}")
print(f"Jumlah mahasiswa perempuan: {jmlP.size/10}")
jmlTI = df[df['Jurusan']=='TI']
jmlTS = df[df['Jurusan']=='TS']
jmlTE = df[df['Jurusan']=='TE']
jmlTM = df[df['Jurusan']=='TM']
print(f"Jumlah mahasiswa jurusan Teknik Informatika: {jmlTI.size/10}")
print(f"Jumlah mahasiswa jurusan Teknik Sipil      : {jmlTS.size/10}")
print(f"Jumlah mahasiswa jurusan Teknik Elektro    : {jmlTE.size/10}")
print(f"Jumlah mahasiswa jurusan Teknik Mesin      : {jmlTM.size/10}")
jmlSMA = df[df['Asal Sekolah']=='SMA']
jmlSMK = df[df['Asal Sekolah']=='SMK']
jmlMA = df[df['Asal Sekolah']=='MA']
print(f"Jumlah asal sekolah SMA: {jmlSMA.size/10}")
print(f"Jumlah asal sekolah SMk: {jmlSMK.size/10}")
print(f"Jumlah asal sekolah MA : {jmlMA.size/10}")
smstrTua = df[df['Semester']>=7]
print(f"Jumlah mahasiswa semester tua: {smstrTua.size}")
reguler = df[df['Kelas']=='Reguler']
karyawan = df[df['Kelas']=='Karyawan']
blended = df[df['Kelas']=='Blended']
print(f"Jumlah mahasiswa kelas reguler: {reguler.size/10}")
print(f"Jumlah mahasiswa kelas karyawan: {karyawan.size/10}")
print(f"Jumlah mahasiswa kelas blended: {blended.size/10}")
mtk = df[df['Mata Kuliah']=='MTK']
print(f"Jumlah mahasiswa yang belajar matematika: {mtk.size/10}")
absen = df[df['Absen']=='Hadir']
print(f"Jumlah absen yang hadir dikelas: {absen.size/10}")
print(f"Rata-rata IPK: {df.IPK.mean()}")
bea = df[df['Kategori']=='Beasiswa']
non = df[df['Kategori']=='Non Beasiswa']
print(f"Jumlah mahasiswa beasiswa: {bea.size/10}")
print(f"Jumlah mahasiswa non-beasiswa: {non.size/10}")

df.to_csv('dataMahasiswa.CSV')
```
