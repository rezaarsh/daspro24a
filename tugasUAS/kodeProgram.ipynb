```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('50_Startups.csv')
df
```

```python
#Statistik Dasar
df.describe()
```

```python
# Visualisasi Statistik Dasar
desc = df.describe()

desc.plot(kind='bar', figsize=(10,5))
plt.title("Statistik Deskriptif")
plt.ylabel("Nilai")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
x = df[['R&D Spend']]
y = df['Profit']

model = LinearRegression()
model.fit(x,y)

inputBaru = pd.DataFrame({'R&D Spend': [150000.0]})

prediksi = model.predict(inputBaru)[0]
print(f"Hasil Prediksi    : {prediksi}\n")

print("Koefisien Regresi :", model.coef_[0])
print("Intercept         :", model.intercept_)
print("R-squared (R2)    :", model.score(x,y))
```

```python
pred = model.predict(x)

plt.figure(figsize=(10,5))
sns.scatterplot(x=df['R&D Spend'], y=df['Profit'], label='Data Asli') 
plt.plot(df['R&D Spend'], pred, color='black', label='Garis Regresi')

plt.scatter(inputBaru, prediksi, s=100, label='Prediksi Input Baru', zorder=5)

plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.title("Prediksi Profit berdasarkan R&D Spend")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
